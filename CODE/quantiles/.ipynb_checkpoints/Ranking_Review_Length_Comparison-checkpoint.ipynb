The purpose of this notebook will be to explore the relationship between the length of reviews, and the number of stars given in a review.  If there appears to be a positive correlation, then it may make sense to use review length as a "tie breaker" when it comes to determinig the absolute raking of products based on user "stars."

For statistical purposes, we will begin with the following null hypothesis, and hope to reject it.

H_o: Higher product ratings (stars) are not correlated with the length of review text.

The following code will import the required libraries and load the data.

```python
import pandas as pd
import sqlite3
import numpy as np
from matplotlib import pyplot as plt

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("../pets.db")
df_raw = pd.read_sql_query("SELECT * from reviews", con)
con.close()
print(df_raw.head())
print("Data Loaded.")
```

In order to reduce the number of "fake" reviews, the following code will limit test dataset to just those reviews that are verified and have a length betwenn 10 and 1000 characters.  Histograms are displayed to give an idea of the distribution of the length of review text as well as the number of stars given as part of reviews.

```python
df_length = df_raw.copy()
df_length = df_length[df_length["verified"]==1]

df_length['length'] = df_raw.apply(lambda row: len(row['review_text']), axis=1)
df = df_length[df_length['length']>10]
df = df[df['length']<1000]

df = df[['overall','length']]

plt.hist(df['overall'])
plt.gca().set_title("Histogram of product ratings (stars)")
plt.show()
plt.hist(df['length'])
plt.gca().set_title("Histogram of product review text length (characters)")
plt.show()
```

Again, the goal of this experiment is to look at how various statistics of the number of reviews across the different ranking levels.  The following will group the reviews by ranking (stars), and compute the sum, mean, median and standard deviation of the aggregated data.  

```python
df_sum_length = df.groupby('overall').sum()
df_mean_length = df.groupby('overall').mean()
df_stdev_length = df.groupby('overall').std()
df_median_length = df.groupby('overall').median()

print("Total sum of review text length by stars.")
plt.bar(df_mean_length.index, df_sum_length['length'])
plt.gca().set_xlabel("Stars")
plt.gca().set_ylabel("Length")
plt.show()

print("Mean of review text length by stars.")
plt.bar(df_mean_length.index, df_mean_length['length'])
plt.gca().set_xlabel("Stars")
plt.gca().set_ylabel("Length")
plt.show()

print("Stdev of review text length by stars.")
plt.bar(df_stdev_length.index, df_stdev_length['length'])
plt.gca().set_xlabel("Stars")
plt.gca().set_ylabel("Length Stdev")
plt.show()

print("Median of review text length by stars.")
plt.bar(df_median_length.index, df_median_length['length'])
plt.gca().set_xlabel("Stars")
plt.gca().set_ylabel("Length")
plt.show()
```

At this most basic level, it does not appear that overall ranking is even remotely correlated with the number of stars.  The only possible exception might be the sum of the characters, however that number is artifically inflated due to the fact that the rakings are heavily skewed to the left.  (In other words, most products receive 5 stars.)
